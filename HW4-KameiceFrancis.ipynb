```python
import geopandas as gpd
```

```python
Feb21_base = gpd.read_file(r"C:\Users\caree\Box\1_HW\Data\Boston_AirBnB_Feb2021\neighbourhoods.geojson")
Feb21_base
```

```python
fig, ax = plt.subplots(figsize=(16,16))
Feb21_base.plot(ax=ax)
list_gdf = gpd.GeoDataFrame(listings_hw21, geometry=gpd.points_from_xy(listings_hw21.longitude, listings_hw21.latitude))
list_gdf.plot(ax=ax, legend=True, color="blue", edgecolor="grey", linewidth=0.2,) 
plt.title('AirBnB Listings in Boston-February 2021', fontsize=40, fontname= "Palatino Linotype", color="blue")
ax.axis("on") 
plt.axis('equal') 
plt.show()

```
